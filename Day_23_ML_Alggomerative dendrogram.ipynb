```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('segmented_customers.xls')
df.head()
```

```python
df. drop('cluster',axis=1,inplace=True) 
df.head()
```

```python
data = df.iloc[:,3:].values
data
```

```python
import scipy.cluster.hierarchy as sch
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(10,7))
plt.title('customer dendogram')
dendo = sch.dendrogram(sch.linkage(data,method = 'ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

clust = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage='ward')
label = clust.fit_predict(data)
```

```python
label
```

```python
plt.figure(figsize=(15,8))
plt.scatter(data[:,0],data[:,1], c = clust.labels_ , cmap='rainbow')
plt.title('customer dendogram')
plt.xlabel('Spending Score')
plt.ylabel('Annual Income')
```
