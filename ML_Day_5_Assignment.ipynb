# Working With DateTime

```python
data['date'] = pd.to_datetime(data['date'])
data.info()
```

```python
data['day'] = data['date'].dt.day
data['month'] = data['date'].dt.month
data['year'] = data['date'].dt.year
data['dayofweek'] = data['date'].dt.dayofweek
data.head()
```

```python
today = pd.to_datetime('today')
today = today.round(freq = 's')
print(today)
```

```python
data['total_time_spend'] = today - data['date']
data.head()
```

```python
data['weekend'] = np.where(data['date'].dt.dayofweek > 5, 1, 0)
data.head()
```

```python
data['WeekDayName']=data['date'].dt.day_name()
data['month']=np.where(data['date'].dt.day_name().isin(['Saturday', 'Sunday']),1,0)
data
```
